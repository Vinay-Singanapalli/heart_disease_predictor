```python
import pandas as pd
```

```python
df=pd.read_csv("heart_data.csv")
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
X=df.drop(columns=["heart disease"])
Y=df["heart disease"]
```

```python
X.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train_scaled=ss.fit_transform(X_train)
X_test_scaled=ss.transform(X_test)
```

```python
X_train_scaled
```

```python
new_x_train=pd.DataFrame(X_train_scaled,columns=X.columns)
```

```python
new_x_train
```

```python
new_x_test=pd.DataFrame(X_test_scaled,columns=X.columns)
```

```python
new_x_test.head(5)
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(new_x_train,Y_train)
```

```python
Y_pred=log.predict(new_x_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy = accuracy_score(Y_test,Y_pred)
precision = precision_score(Y_test,Y_pred)
recall= recall_score(Y_test,Y_pred)
f1=f1_score(Y_test,Y_pred)
print(f"accuracy : {accuracy}  ,  precision : {precision} , recall : {recall} , f1_score : {f1}")
```

```python
import joblib
```

```python
joblib.dump(log,"heart.pkl")
```
